```python
with open('output.txt', 'r') as input:
    text = input.read().split('\n')
text
```

```python
import matplotlib.pyplot as plt

global_index = 0
cnt_devices = 3
cnt_groups = 6
cnt_n = 9

f, axs = plt.subplots(3,2,figsize=(15,5))
index_axs = 0
for device in range(cnt_devices):
    device_name = text[global_index]
    global_index += 1
    
    n_list = list()
    float_groups = list()
    for group in range(cnt_groups):
        one_group = list()
        for n in range(cnt_n):
            one_group.append(float(text[global_index].split()[6]))
            global_index += 1
        float_groups.append(one_group)
        
    double_groups = list()
    for group in range(cnt_groups):
        one_group = list()
        n_list = list()
        for n in range(cnt_n):
            one_group.append(float(text[global_index].split()[6]))
            n_list.append(int(text[global_index].split()[2]))
            global_index += 1
        double_groups.append(one_group)

    
    axs[0][0].plot(
        n_list, float_groups[0], 'red',
        n_list, float_groups[1], 'green',
        n_list, float_groups[2], 'yellow',
        n_list, float_groups[3], 'purple',
        n_list, float_groups[4], 'blue',
        n_list, float_groups[5], 'black',
    )
    axs[0][0].legend(['group 8', 'group 16', 'group 32', 'group 64', 'group 128', 'group 256'])
    axs[0][0].set_title(device_name + ' on float')
    axs[0][0].set_xlabel('N')
    axs[0][0].set_ylabel('Time')
    index_axs += 1
    
    axs[0][0].plot(
        n_list, float_groups[0], 'red',
        n_list, float_groups[1], 'green',
        n_list, float_groups[2], 'yellow',
        n_list, float_groups[3], 'purple',
        n_list, float_groups[4], 'blue',
        n_list, float_groups[5], 'black',
    )
    axs[0][0].legend(['group 8', 'group 16', 'group 32', 'group 64', 'group 128', 'group 256'])
    axs[0][0].set_title(device_name + ' on double')
    axs[0][0].set_xlabel('N')
    axs[0][0].set_ylabel('Time')
    index_axs += 1


plt.show()
    
```
